# Loading Data....

```python
data=pd.read_csv('HireArt - Data Analyst Exercise 10.12.17 - Sheet1.csv')
data.head()
```

```python
#Count the numbers of Account manager
data2=data.groupby(['Account manager'])
data2.size()
```

```python
#use pivot table to format data for each Account manager
data3=data.pivot_table(index='Account manager', columns='Date of Contact', values='Client Name', aggfunc='count')
data3
```

# Data Analysis and Visualization

## 1. Split out the contacts made by Account Manager: Milan Crona, name it as data_MilanCrona

```python
data_MilanCrona=data.groupby(['Account manager']).get_group('Milan Crona')
data_MilanCrona.head()
# data_Crona=data4.pivot_table(index='Date of Contact', columns='Account manager', values='Client Name', aggfunc='count')
# data_Crona.sort_values('Milan Crona',ascending=False)
```

```python
# Convert that column into a datetime datatype
data_MilanCrona['Date of Contact'] = pd.to_datetime(data_MilanCrona['Date of Contact'])
# Set the datetime column as the index
data_MilanCrona.index = data_MilanCrona['Date of Contact']
data_MilanCrona=data_MilanCrona.drop(["Account manager","Client Name"], axis=1)
data_MilanCrona_1=data_MilanCrona.resample('M').count()
data_MilanCrona_1.head()
```

```python
# data5.plot.bar(figsize=(20, 8))
```

```python
data6=data_MilanCrona_1.sort_values('Date of Contact', ascending=True)

data6.plot.barh(figsize=(10,15), title='Contacts made by Account Manager: Milan Crona')
```
